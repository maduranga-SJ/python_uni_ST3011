```python
# BAR GRPH 


'''
import matplotlib
matplotlib.pyplot.plot(...)

------------------------------------

import pylab
pylab.plot(...)

both of above imports results same 
'''

import matplotlib.pyplot as plt; 
plt.rcdefaults() # restore the pyplot default parameters
    # pyplot for ploting / plot(), bar()
import numpy as np 
    # numpy - for multi dimensional arrays// arange()
import matplotlib.pyplot as plt

# x - variable
objects = ('Python','C++','JAVA','Perl','Scala','Lisp')
y_pos = np.arange(len(objects))
'''
y =np.arange(6)
print(y)
-----
[0 1 2 3 4 5]
'''

#Y- variable 
performance = [10,8,6,4,2,1]

#build bar chart 
plt.bar(y_pos,performance,align = 'center', alpha = 0.5)
'''
align - [center / edge ] - alignment of bars with the x variables 
alpha - opacity 0.0 - 1
'''

# define categories for x values
plt.xticks(y_pos,objects)

#labels
plt.ylabel("performance")
plt.xlabel("languange")
plt.title("Performance of languages ")

```

```python
# MULTIPLE BAR CHARTS 
#typical imports >>>>
import numpy as np
import matplotlib.pyplot as plt

# DATA
n_groups = 4 # number of groups 
means_frank = (90,55,40,65) 
# Tuples [cannot change the values after assignment ] List s eg :- a = [1,2,3] can change te values
means_guido= (85,62,54,20)

#CREATE PLOT
fig , ax = plt.subplots()# not necessary ??? (*)f
index = np.arange(n_groups)# number of x variables
bar_width = 0.35 # to define the bar positions of the sub plot 2
opacity = 0.8

sub1 = plt.bar(index,means_frank,bar_width,alpha = opacity , color = 'r', label = 'Frank')
sub2 = plt.bar(index + bar_width , means_guido,bar_width, alpha = opacity , color = 'y', label = 'Quido')
#  adjust the bar location ^                   ^ width is needed to be defined to adjust the bar 

#label 
plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by Person')
plt.xticks(index + bar_width,('A','B','C','D'))
plt.legend()# show legend (colors vs sub plot)

plt.tight_layout()
plt.show()
```
